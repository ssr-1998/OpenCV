# Image Similarity

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from skimage.metrics import structural_similarity as SSIM
```

## Mean Squared Error :-
![title](Images/Mean_Squared_Error.png "Header")<br>
This formula might look complex but let's understand it in Python code because in that way it will be alot simpler.

```python
# The 'Mean Squared Error' between the two images is the sum of the squared difference between the two images.
# NOTE: the two images must have the same dimension

def MSE(image1, image2):
        error = np.sum((image1.astype("float") - image2.astype("float"))**2)
        error /= float(image1.shape[0] * image1.shape[1])
        return error
```

Let’s tear the above function and see what’s going on:<br>
First, We define our MSE function, which takes two arguments: image1 & image2 (i.e. the images we want to compare for similarity).<br>
All the real work is handled in second line. First we convert the images from unsigned 8-bit integers to floating point, that way we don’t run into any problems with modulus operations “wrapping around”. We then take the difference between the images by subtracting the pixel intensities. Next up, we square these differences (hence mean squared error) and finally sum them up.
In the next line of code we handles the mean of the Mean Squared Error. All we are doing is dividing our sum of squares by the total number of pixels in the image.<br>
Finally, we return our MSE to the caller.<br>
MSE is dead simple to implement — but when using it for similarity, we can run into problems. The main one being that large distances between pixel intensities do not necessarily mean the contents of the images which are dramatically different.<br><br>
**Note :-** A value of 0 for MSE indicates perfect similarity. A value greater than one implies less similarity and will continue to grow as the average difference between pixel intensities increases as well.

## Structural Similarity Index :-
Luckily, we don't need to implement this method by hand, as sikit-image has already implemented it. We can import it by following command :-<br>
**from skimage.metrics import structural_similarity as SSIM**<br><br>
**Note :-** Unlike MSE, the SSIM value can vary between -1 and 1, where 1 indicates perfect similarity.
## Comparing Image Similarity by Both :-

```python
def Compare_Images(image1, image2):
    mse = MSE(image1, image2)
    ssim = SSIM(image1, image2)
    
    fig, (ax1,ax2) = plt.subplots(1, 2, figsize=(20,20))

    ax1.imshow(image1, cmap = "gray")
    ax1.axis("off")

    ax2.imshow(image2, cmap = "gray")
    ax2.axis("off")

    print("MSE: %.2f, SSIM: %.2f" % (mse, ssim))
    plt.show()
```

### Creating Sample Images :-

```python
faceR = cv2.imread("Images/faceR.jpeg", 0)
faceRI = cv2.imread("Images/faceRI.png", 0)
faceRN5 = cv2.imread("Images/faceRN5.png", 0)

print(faceR.shape)
print(faceRI.shape)
print(faceRN5.shape)
print("X"+"--X"*40)

faceR = faceR[0:470, 0:552]
faceRI = faceRI[0:470, 0:552]

print(faceR.shape)
print(faceRI.shape)
print(faceRN5.shape)

fig = plt.figure(figsize=(20,20))

ax = fig.add_subplot(1,3,1).set_title("Original Image")
plt.imshow(faceR, cmap="gray")

ax = fig.add_subplot(1,3,2).set_title("Enhanced Image")
plt.imshow(faceRI, cmap="gray")

ax = fig.add_subplot(1,3,3).set_title("Noised Image")
plt.imshow(faceRN5, cmap="gray")

plt.show()
```

## Original Vs Original :-

```python
Compare_Images(faceR, faceR)
```

## Original Vs Enchanced Image :-

```python
Compare_Images(faceR, faceRI)
```

## Original Vs Noised Image :-

```python
Compare_Images(faceR, faceRN5)
```

**So, we can see that SSIM performs better than MSE for Image Similarity.**
