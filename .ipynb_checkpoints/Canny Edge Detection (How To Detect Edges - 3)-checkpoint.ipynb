**Canny() Function :-**<br>
It is a function in which we will set lower & upper threshold according to our choice i.e. if we want we can keep the gap of thresholds wide or can tight the gap of thresholds. When wide the function will be able to count more high frequency points in the image and will accordingly provide us a image with high intensity of edges whereas when tight the function won't be able to count more high frequency points though won't be giving us high intensity of edges.<br>
This is a function which will automatically 1st apply **Low Pass Filtering** by bluring and then will perform **High Pass Filtering** process but in this it will combine sobel_x & sobel_y both to get better results but this will also increase the pixels. Then by performing the last step i.e. is **Thresholding** it gives us a beautiful edged image of object.

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from matplotlib.pyplot import rcParams

%matplotlib inline
rcParams["figure.figsize"] = 10,10
```

```python
# Read in the image
image = cv2.imread('images/brain_MR.jpg')

# Change color to RGB (from BGR)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Convert the image to grayscale for processing
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(20,20))
ax2.imshow(gray, cmap='gray')
ax2.set_title("Gray Image")
ax1.imshow(image)
ax1.set_title("Original Image")
plt.show()
```

### Implement Canny edge detection

```python
# Try Canny using "wide" and "tight" thresholds
"""If there is wide gap between thresholds then it will capture more edges whereas when gap is tight there will be low edges."""
wide = cv2.Canny(gray, 30, 100)
tight = cv2.Canny(gray, 200, 240)
 
    
# Display the images
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))

ax1.set_title('wide')
ax1.imshow(wide, cmap='gray')

ax2.set_title('tight')
ax2.imshow(tight, cmap='gray')
plt.show()
```

### TODO: Try to find the edges of this flower

Set a small enough threshold to isolate the boundary of the flower.

```python
# Read in the image
image = cv2.imread('images/sunflower.jpg')

# Change color to RGB (from BGR)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

fig, (ax1,ax2) = plt.subplots(1,2,figsize=(20,20))
ax1.imshow(image)
ax1.set_title("Original Flower Image")
ax2.imshow(gray, cmap="gray")
ax2.set_title("Gray Image")
plt.show()
```

```python
## TODO: Define lower and upper thresholds for hysteresis
# right now the threshold is so small and low that it will pick up a lot of noise
# lower = 0
# upper = 50
edges_wide = cv2.Canny(gray, 150, 300)
edges_tight = cv2.Canny(gray, 20,70)
fig, (ax1, ax2) = plt.subplots(1,2,figsize=(20,20))
ax1.imshow(edges_wide, cmap='gray')
ax1.set_title("Edged Flower Image {Wide}")
ax2.imshow(edges_tight, cmap="gray")
ax2.set_title("Edged Flower Image {Tight}")
plt.show()
```

## Let's try some other pics too :-


```python
car = cv2.imread("images/car.jpeg")
car = cv2.cvtColor(car, cv2.COLOR_BGR2RGB)
gray_car = cv2.cvtColor(car, cv2.COLOR_RGB2GRAY)

fig, (ax1,ax2) = plt.subplots(1,2, figsize=(20,20))

ax1.imshow(car)
ax1.set_title("Original Image of Car")
ax2.imshow(gray_car, cmap="gray")
ax2.set_title("Gray Car")
plt.show()
```

```python
car_edges_wide = cv2.Canny(gray_car, 50,500)
car_edges_tight = cv2.Canny(gray_car, 100,105)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,20))
ax1.imshow(car_edges_wide, cmap="gray")
ax1.set_title("Edged Image Of Car {Wide}")
ax2.imshow(car_edges_tight, cmap="gray")
ax2.set_title("Edged Image Of Car {Tight}")
plt.show()
```
