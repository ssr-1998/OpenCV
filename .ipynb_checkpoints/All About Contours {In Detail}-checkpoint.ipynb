# Contours :-
Contours can be explained simply as a curve joining all the continuous points (along the boundary), having same color or intensity. The contours are a useful tool for shape analysis and object detection and recognition.

````python
# Loading Librarires :-
import numpy as np
import matplotlib.pyplot as plt
import cv2

# Loading Image :-
square = cv2.imread("Images/Square_1.png")
square = cv2.cvtColor(square, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(square, cv2.COLOR_BGR2GRAY)

print("Height - {} \nWidth - {}".format(gray.shape[0], gray.shape[1]))
plt.imshow(gray, cmap="gray")
plt.show()
````

````python
# Thresholding the image as it is better to put background black & foreground white by using Thresholding.
thresh = cv2.threshold(gray,150, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]
plt.imshow(thresh, cmap="gray")
plt.show()
````

### Contour Retrieval Mode
This argument is jsut responsible for the Hierarchy(relation) which is among Contours like which Contour is Parent or which one is child & even this that which one is on the same level of Parent or Child (Think of a Square inside a Circle on a black background, Circle Contour will be named as Parent & Child will be the Contour of Square). Hierarchy for each Contour is returned in  **[Next, Previous, First_Child, Parent]** format. Next & Previous tells about the Index Numbers of Next & Previous Contours which are on that same level if any otherwise will put (-1). In the same way it puts Child & Parents Index Numbers & if there are no Child / Parent then will put (-1).

**Best Retrieval Modes are :-**<br>
**1.  cv2.RETR_LIST ->** This simply generates all the Contours but doesn't create any Child-Parent Relationship.<br>
**2. cv2.RETR_TREE ->** This also generates all the Contours but unlike above one, this creates all the possible relationships among Contours (will even tell you about Grand Father & Grand Sons also). **:)**<br>

**To know more about Contour Retrieval Mode see this ->** https://opencv-python-tutroals.readthedocs.io/en/latest/py_tutorials/py_imgproc/py_contours/py_contours_hierarchy/py_contours_hierarchy.html

### Contour Approximation Method
This is the third argument in **cv2.findContours** function.

Above, I told that contours are the boundaries of a shape with same intensity. It stores the (x,y) coordinates of the boundary of a shape. But does it store all the coordinates ? That is specified by this contour approximation method.

If you pass **cv2.CHAIN_APPROX_NONE**, all the boundary points are stored. But actually do we need all the points? For eg, you found the contour of a straight line. Do you need all the points on the line to represent that line? No, we need just two end points of that line. This is what **cv2.CHAIN_APPROX_SIMPLE** does. It removes all redundant points and compresses the contour, thereby saving memory.

````python
"""cv2.findCountorus() function returns 3 things.
1. Image array, the same array which we supply to it.
2. Countours Coordinates
3. Hierarchy"""

# cv2.CHAIN_APPROX_None ->
conts_by_none = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)[1]

# cv2.CHAIN_APPROX_Simple ->
conts_by_simple = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)[1]
````

Let's give a look at Contours

In above 2 cells :-
1st Cell shows the Contours gained from **cv2.CHAIN_APPROX_NONE** & 2nd shows the Contours gained from **cv2.CHAIN_APPROX_SIMPLE.** See, how much memory it saves!!!

Now, will draw the countours on **Original Image**. But we have to make copies of Original Image as **cv2.drawContours()** function modifies the Original Image.

````python
# Drawing Contours of cv2.CHAIN_APPROX_None ->
square_1 = square.copy()
conts_by_none_draw = cv2.drawContours(square_1, conts_by_none, -1, (255, 0, 0), 2)  # -1 is passed to draw all countours.

# Drawing Contours of cv2.CHAIN_APPROX_Simple ->
square_2 = square.copy()
conts_by_simple_draw = cv2.drawContours(square_2, conts_by_simple, -1, (0, 0, 255), 2)

# Plotting Result Images
f, (ax1, ax2) = plt.subplots(1, 2 ,figsize=(10,15))
ax1.imshow(conts_by_none_draw)
ax1.set_title("None")

ax2.imshow(conts_by_simple_draw)
ax2.set_title("Simple")

plt.show()
````

After Drawing Contours by both Approximation Methods we can see that there is no difference. So, it is better to use **cv2.CHAIN_APPROX_SIMPLE.**

### Import resources and display image

````python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline

# Read in the image
image = cv2.imread('images/thumbs_up_down.jpg')

# Change color to RGB (from BGR)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
````

### Produce a binary image for finding contours

````python
# Convert to grayscale
gray = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)

# Create a binary thresholded image
thresh = cv2.threshold(gray, 225, 255, cv2.THRESH_BINARY_INV)[1]

plt.imshow(thresh, cmap='gray')
````

### Find and draw the contours

````python
# Find contours from thresholded, binary image
retval, contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# Draw all contours on a copy of the original image
contours_image = np.copy(image)
contours_image = cv2.drawContours(contours_image, contours, -1, (0,255,0), 3)  # -1 means to draw all contours & thickness = 3.

plt.imshow(contours_image)
plt.show()
````

## Contour Features

Every contour has a number of features that you can calculate, including the area of the contour, it's orientation (the direction that most of the contour is pointing in), it's perimeter, and many other properties outlined in [OpenCV documentation, here](http://opencv-python-tutroals.readthedocs.io/en/latest/py_tutorials/py_imgproc/py_contours/py_contour_properties/py_contour_properties.html).

In the next cell, you'll be asked to identify the orientations of both the left and right hand contours. The orientation should give you an idea of which hand has its thumb up and which one has its thumb down!

### Orientation

The orientation of an object is the angle at which an object is directed. To find the angle of a contour, you should first find an ellipse that fits the contour and then extract the `angle` from that shape.
 
```python
# Fit an ellipse to a contour and extract the angle from that ellipse
(x,y), (MA,ma), angle = cv2.fitEllipse(selected_contour)
```

**Orientation values**

These orientation values are in degrees measured from the x-axis. A value of zero means a flat line, and a value of 90 means that a contour is pointing straight up!

So, the orientation angles that you calculated for each contour should be able to tell us something about the general position of the hand. The hand with it's thumb up, should have a higher (closer to 90 degrees) orientation than the hand with it's thumb down.

### TODO: Find the orientation of each contour

````python
## TODO: Complete this function so that 
## it returns the orientations of a list of contours
## The list should be in the same order as the contours
## i.e. the first angle should be the orientation of the first contour
def orientations(contours):
    """
    Orientation 
    :param contours: a list of contours
    :return: angles, the orientations of the contours
    """
    
    # Create an empty list to store the angles in
    # Tip: Use angles.append(value) to add values to this list
    angles = []
    for i in range(0, len(contours)):
        (x,y), (MA,ma), angle = cv2.fitEllipse(contours[i])  # MA -> Major Axis & ma -> Minor Axis
        angles.append(angle)
    return angles


# ---------------------------------------------------------- #
# Print out the orientation values
angles = orientations(contours)
print('Angles of each contour (in degrees): ' + str(angles))
````
