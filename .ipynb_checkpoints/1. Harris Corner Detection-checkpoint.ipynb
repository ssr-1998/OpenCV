# Harris Corner Detection

### Import resources and display image

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

%matplotlib inline

# Read in the image
image = cv2.imread('images/waffle.jpg')

# Make a copy of the image
image_copy = np.copy(image)

# Change color to RGB (from BGR)
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

plt.imshow(image_copy)
```

### Detect corners

```python
# Convert to grayscale
gray = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)
gray = np.float32(gray)

# Detect corners 
dst = cv2.cornerHarris(gray, 2, 3, 0.04)

# Dilate corner image to enhance corner points
dst = cv2.dilate(dst,None)

plt.imshow(dst, cmap='gray')
```

### Extract and display strong corners

```python
## TODO: Define a threshold for extracting strong corners
# This value vary depending on the image and how many corners you want to detect
# Try changing this free parameter, 0.1, to be larger or smaller ans see what happens
thresh = 0.1*dst.max()

# Create an image copy to draw corners on
corner_image = np.copy(image_copy)

# Iterate through all the corners and draw them on the image (if they pass the threshold)
for j in range(0, dst.shape[0]):
    for i in range(0, dst.shape[1]):
        if(dst[j,i] > thresh):
            # image, center pt, radius, color, thickness
            cv2.circle( corner_image, (i, j), 1, (0,255,0), 1)

plt.imshow(corner_image)
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

car = cv2.imread("images/mc-laren.jpg")
car = cv2.cvtColor(car, cv2.COLOR_BGR2RGB)

gray_car = cv2.cvtColor(car, cv2.COLOR_RGB2GRAY)

fig, ax = plt.subplots(1,2, figsize=(20,10))
ax[0].imshow(car)
ax[0].set_title("Original Image")
ax[1].imshow(gray_car, cmap="gray")
ax[1].set_title("Gray Image")
plt.show()
```

```python
gray_car = np.float32(gray_car)

# Detect Corners(Here Ksize must be odd & less then 31.)
dst = cv2.cornerHarris(gray_car, 5, 13, 0.04)


# Dilate corner image to enhance corner points
dst = cv2.dilate(dst,None)

plt.imshow(dst, cmap='gray')
plt.show()
```

```python
## TODO: Define a threshold for extracting strong corners
# This value vary depending on the image and how many corners you want to detect
# Try changing this free parameter, 0.1, to be larger or smaller ans see what happens
thresh = 0.01*dst.max()

# Create an image copy to draw corners on
corner_image = np.copy(car)

# Iterate through all the corners and draw them on the image (if they pass the threshold)
for j in range(0, dst.shape[0]):
    for i in range(0, dst.shape[1]):
        if(dst[j,i] > thresh):
            # image, center pt, radius, color, thickness
            cv2.circle( corner_image, (i, j), 1, (0,255,0), 1)

plt.imshow(corner_image)
plt.show()
```

# Shi-Tomasi Corner Detection :-
Better than **Har**ris Corner Detector** as its formula has modifications to find **best N number of corners.** To apply this we will use **cv2.goodFeaturesToTrack()** function & in this we can specify many parameters. If we want to use **Harris Corner Detector** by this as well & we can even do that as it is having a parameter **useHarrisDetector=True/False.** Other than this we can specify the **number of corners** we want to detect & also we can use a **qualityLevel** parameter too, which takes values between **0 to 1.** In whole image whichever corner will have best quality it will multiply the value of **qualityLevel** from that highest quality and the output will be seen as a **threshold** according to which only the corners which will have higher value than that threshold will only be suplied to us otherwise will get rejected.<br>
**For Example :-** Suppose in an image a corner's highest quality is 1500 and we set the qualityLevel parameter value as 0.01. Then the function will reject all the corners which will be having quality value less than 15, & only with qualtiy over it will be supplied to us.

```python
img = cv2.imread("images/mc-laren.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.imread("images/mc-laren.jpg", 0)

corners = cv2.goodFeaturesToTrack(gray, maxCorners=724 ,qualityLevel=0.01, minDistance=5, useHarrisDetector=False)
print("There are",len(corners), "number of corners")
for i in corners:
    x,y = i.ravel()
    cv2.circle(img,(x,y),3,255,2)

plt.imshow(img)
plt.show()
```
