## Image pyramids {Gaussian Pyramid Technique}

**Note :- Low Resolution Images means the images which have low pixels or can say those pictures have low quality, whereas High Resolution Images have high quality.**<br><br>

Normally, we used to work with an image of constant size. But in some occassions, we need to work with images of different resolution of the same image. For example, while searching for something in an image, like face, we are not sure at what size the object will be present in the image. In that case, we will need to create a set of images with different resolution and search for object in all the images. These set of images with different resolution are called Image Pyramids (because when they are kept in a stack with biggest image at bottom and smallest image at top look like a pyramid).

Higher level (Low resolution) in a Gaussian Pyramid is formed by removing consecutive rows and columns in Lower level (higher resolution) image. Then each pixel in Low Resolution is formed by the contribution from 5 pixels in underlying level with gaussian weights. By doing so, a M \times N image becomes M/2 \times N/2 image. So area reduces to one-fourth of original area. It is called an Octave. We can find Gaussian pyramids using cv2.pyrDown() and cv2.pyrUp() functions.

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

# Read in the image
image = cv2.imread('images/rainbow_flag.jpg')

# Change color to RGB (from BGR)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

print("Height (y-axis max value) :- {}\n Width (x-axis max value) :- {}\n & Channels :- {}".format(
    image.shape[0], image.shape[1], image.shape[2]))

plt.imshow(image)
plt.show()
```

**cv2.pyrDown() ->**

```python
# Here, I will first use cv2.pyrDown() to covert high resolution image to low resolution image. Taking Level = 3.
down_level_1 = cv2.pyrDown(image)
down_level_2 = cv2.pyrDown(down_level_1)
down_level_3 = cv2.pyrDown(down_level_2)

# Display the images
f, (ax1,ax2,ax3,ax4) = plt.subplots(1, 4, figsize=(20,15))

ax1.set_title('Original')
ax1.imshow(image)

ax2.set_title('Down Level - 1')
ax2.imshow(down_level_1)
ax2.set_xlim([0, image.shape[1]])
ax2.set_ylim([0, image.shape[0]])

ax3.set_title('Down Level - 2')
ax3.imshow(down_level_2)
ax3.set_xlim([0, image.shape[1]])
ax3.set_ylim([0, image.shape[0]])

ax4.set_title('Down Level - 3')
ax4.imshow(down_level_3)
ax4.set_xlim([0, image.shape[1]])
ax4.set_ylim([0, image.shape[0]])
plt.show()
```

**cv2.pyrUp() ->**

```python
up_level_1 = cv2.pyrUp(down_level_3)
up_level_2 = cv2.pyrUp(up_level_1)
up_level_3 = cv2.pyrUp(up_level_2)

fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(20,15))

ax1.set_title('Down Level - 3')
ax1.set_xlim([0, image.shape[1]])
ax1.set_ylim([0, image.shape[0]])
ax1.imshow(down_level_3)

ax2.set_title('Up Level - 1')
ax2.imshow(up_level_1)
ax2.set_xlim([0, image.shape[1]])
ax2.set_ylim([0, image.shape[0]])

ax3.set_title('Up Level - 2')
ax3.imshow(up_level_2)
ax3.set_xlim([0, image.shape[1]])
ax3.set_ylim([0, image.shape[0]])

ax4.set_title('Up Level - 3')
ax4.imshow(up_level_3)
ax4.set_xlim([0, image.shape[1]])
ax4.set_ylim([0, image.shape[0]])
plt.show()
```

Here, we saw that we converted an High Resolution Image to a Low Resolution Image while applying Image Miniaturizing Techniques. But again when we Normalized that image by using cv2.pyrUp(), it didn't restored the Image Resolution. So, we know this now that after applying Image Pyramids we we loose information from images.
