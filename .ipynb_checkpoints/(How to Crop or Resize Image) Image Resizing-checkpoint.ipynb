```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

## 1st Example :-

```python
# Loading Image
image_BGR = cv2.imread("Images/waymo_car.jpg")
image_RGB = cv2.cvtColor(image_BGR, cv2.COLOR_BGR2RGB)
print("Height (y-axis max value) :- {}\n Width (x-axis max value) :- {}\n & Channels :- {}".format(
    image_BGR.shape[0], image_BGR.shape[1], image_BGR.shape[2]))
plt.imshow(image_RGB)
cv2.imshow("Original Image", image_BGR)
cv2.waitKey(0)
cv2.destroyAllWindows()
plt.show()
```

## First let's see the Resizing Concept :-
Main difference among Original & Resized Image can be seen by the Images we show by cv2.imshow().

```python
heigth, width = 200, 200
resized_image_RGB = cv2.resize(image_RGB, (heigth, width))
resized_image_BGR = cv2.resize(image_BGR, (heigth, width))
print("Height (y-axis max value) :- {}\n Width (x-axis max value) :- {}\n & Channels :- {}".format(
    resized_image_BGR.shape[0], resized_image_BGR.shape[1], resized_image_BGR.shape[2]))
plt.imshow(resized_image_RGB)
cv2.imshow("Resized Image", resized_image_BGR)
cv2.waitKey(0)
cv2.destroyAllWindows()
plt.show()
```

```python
heigth, width = 1000, 1000
resized_image_RGB = cv2.resize(image_RGB, (heigth, width))
resized_image_BGR = cv2.resize(image_BGR, (heigth, width))
print("Height (y-axis max value) :- {}\n Width (x-axis max value) :- {}\n & Channels :- {}".format(
    resized_image_BGR.shape[0], resized_image_BGR.shape[1], resized_image_BGR.shape[2]))
plt.imshow(resized_image_RGB)
cv2.imshow("Resized Image", resized_image_BGR)
cv2.waitKey(0)
cv2.destroyAllWindows()
plt.show()
```
