# Detecting Edges Of Medical Image :-

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from matplotlib.pyplot import rcParams
%matplotlib inline
rcParams["figure.figsize"] = 10,10
```

Loading brain image & converting it to RGB channel from BGR channel

```python
# Read in the image
brain = cv2.imread('images/brain_MR.jpg')

# Make a copy of the image
brain_copy = np.copy(brain)

# Change color to RGB (from BGR)
brain_copy = cv2.cvtColor(brain_copy, cv2.COLOR_BGR2RGB)

plt.imshow(brain_copy)
plt.show()
```

Now, will convert brain image to **Gray Scale** to easily manipulate & process image to get it's edges.<br><br>
Here we will also apply  **Low Pass Filtering** on image which means to decrease the noise of image which we see when we process the image with **High Pass Filtering**, it means the bright points which are not edges but still remains in image after processing with **High Pass Filtering**.<br>
So, to remove it we use many techniques in **Low Pass Filtering** one of them is **Bluring** the image. This is one technique to remove or decrease the level of noise in image.<br><br>
There are mainly 5 ways to **Blur** but the important ways are only **4** :-<br>
**1.** **cv2.blur() function**<br>
**2.** **cv2.medianBlur() function**<br>
**3.** **cv2.GaussianBlur() function**<br>
**4.** **cv2.bilateralFilter() function**<br><br>
**bilaretalFilter() function** mostly performs **best & then comes** the **GaussianBlur() function**.<br> So, here we will use only these two to **remove noises.**


```python
# Convert to grayscale for filtering
gray_brain = cv2.cvtColor(brain_copy, cv2.COLOR_RGB2GRAY)

# Creatng Gaussian & Bilateral Blurred Image.
gray_gaussian = cv2.GaussianBlur(gray_brain, (9, 9), 0)
gray_bilateral = cv2.bilateralFilter(gray_brain, 10, 75,75)

f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20,10))
ax1.set_title('Original Gray Brain Image')
ax1.imshow(gray_brain, cmap='gray')

ax2.set_title('Blurred Gaussian Brain Image')
ax2.imshow(gray_gaussian, cmap='gray')

ax3.set_title('Blurred Bilateral Brain Image')
ax3.imshow(gray_bilateral, cmap='gray')
plt.show()
```

Now, will use **High Pass Filtering** to get edges 

```python
# High-pass filter 

# 3x3 sobel filters for edge detection
sobel_x = np.array([[ -1, 0, 1], 
                   [ -2, 0, 2], 
                   [ -1, 0, 1]])


sobel_y = np.array([[ -1, -2, -1], 
                   [ 0, 0, 0], 
                   [ 1, 2, 1]])

# Filter the orginal and blurred grayscale images using filter2D
only_sobel_x = cv2.filter2D(gray_brain, -1, sobel_x)
only_sobel_y = cv2.filter2D(gray_brain, -1, sobel_y)
gaussian_sobel_x = cv2.filter2D(gray_gaussian, -1, sobel_x)
gaussian_sobel_y = cv2.filter2D(gray_gaussian, -1, sobel_y)
bilateral_sobel_x = cv2.filter2D(gray_bilateral, -1, sobel_x)
bilateral_sobel_y = cv2.filter2D(gray_bilateral, -1, sobel_y)

f, ax = plt.subplots(3, 2, figsize=(20,20))

ax[0,0].set_title('Only Sobel_x (without Low Pass Filtering)')
ax[0,0].imshow(only_sobel_x, cmap='gray')
ax[0,1].set_title('Only Sobel_y (without Low Pass Filtering)')
ax[0,1].imshow(only_sobel_y, cmap='gray')


ax[1,0].set_title('Gaussian & Sobel_x')
ax[1,0].imshow(gaussian_sobel_x, cmap='gray')
ax[1,1].set_title('Gaussian & Sobel_y')
ax[1,1].imshow(gaussian_sobel_y, cmap='gray')

ax[2,0].set_title('Bilateral & Sobel_x')
ax[2,0].imshow(bilateral_sobel_x, cmap='gray')
ax[2,1].set_title('Bilateral & Sobel_y')
ax[2,1].imshow(bilateral_sobel_y, cmap='gray')
plt.suptitle("Brain Image After High Pass Filtering", fontsize=20)
plt.tight_layout(pad=6)
plt.show()
```

So, as we all can see that there is so much noise in **Only Sobel_x & Only Sobel_y (Without Low Pass Filtering)** image.<br>
Some better performing is **Gaussian & Sobel_y** image but<br>
the best one performing is **Bilateral & Sobel_x** image followed by **Bilateral & Sobel_y** image.

```python
# Create threshold that sets all the filtered pixels to white
# Above a certain threshold

retval_only_sobel_x, binary_only_sobel_x = cv2.threshold(only_sobel_x, 50, 255, cv2.THRESH_BINARY)
retval_only_sobel_y, binary_only_sobel_y = cv2.threshold(only_sobel_y, 50, 255, cv2.THRESH_BINARY)
retval_gaussian_sobel_x, binary_gaussian_sobel_x = cv2.threshold(gaussian_sobel_x, 50, 255, cv2.THRESH_BINARY)
retval_gaussian_sobel_y, binary_gaussian_sobel_y = cv2.threshold(gaussian_sobel_y, 50, 255, cv2.THRESH_BINARY)
retval_bilateral_sobel_x, binary_bilateral_sobel_x = cv2.threshold(bilateral_sobel_x, 50, 255, cv2.THRESH_BINARY)
retval_bilateral_sobel_y, binary_bilateral_sobel_y = cv2.threshold(bilateral_sobel_y, 50, 255, cv2.THRESH_BINARY)

f, ax = plt.subplots(3, 2, figsize=(20,20))

ax[0,0].set_title('Only Sobel_x (without Low Pass Filtering)')
ax[0,0].imshow(binary_only_sobel_x, cmap='gray')
ax[0,1].set_title('Only Sobel_y (without Low Pass Filtering)')
ax[0,1].imshow(binary_only_sobel_y, cmap='gray')


ax[1,0].set_title('Gaussian & Sobel_x')
ax[1,0].imshow(binary_gaussian_sobel_x, cmap='gray')
ax[1,1].set_title('Gaussian & Sobel_y')
ax[1,1].imshow(binary_gaussian_sobel_y, cmap='gray')

ax[2,0].set_title('Bilateral & Sobel_x')
ax[2,0].imshow(binary_bilateral_sobel_x, cmap='gray')
ax[2,1].set_title('Bilateral & Sobel_y')
ax[2,1].imshow(binary_bilateral_sobel_y, cmap='gray')
plt.suptitle("Final Edge Detections", fontsize=20)
plt.tight_layout(pad=6)
plt.show()
```
