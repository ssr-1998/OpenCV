```python
import cv2
import imutils
import matplotlib.pyplot as plt
from skimage.measure import compare_ssim
```

```python
imageA = cv2.imread("Images/Credit_Card.jpg")
imageA = cv2.cvtColor(imageA, cv2.COLOR_BGR2RGB)
imageA = imageA[0:576, 0:496]

imageB = cv2.imread("Images/Credit_Card_2.jpg")
imageB = cv2.cvtColor(imageB, cv2.COLOR_BGR2RGB)

grayA = cv2.cvtColor(imageA, cv2.COLOR_RGB2GRAY)
grayB = cv2.cvtColor(imageB, cv2.COLOR_RGB2GRAY)

fig, ax = plt.subplots(2,2, figsize=(10,10))
ax[0,0].imshow(imageA)
ax[0,0].set_title("Original")
ax[0,1].imshow(imageB)
ax[0,1].set_title("Modified")
ax[1,0].imshow(grayA, cmap="gray")
ax[1,0].set_title("grayA")
ax[1,1].imshow(grayB, cmap="gray")
ax[1,1].set_title("grayB")
plt.show()
```

```python
"""
compare_ssim() gives 2 things in return. 1st is Score & 2nd is Difference. For this task we want the Difference.
Difference(diff) comes in float values as each value is divided by max pixel i.e. 255.
To convert it into image pixels we just multiplied it with 255."""

score, diff = compare_ssim(grayA, grayB, full= True)
print("SSIM Score :-", score)
print("\n Old diff ->",diff)
diff = (diff * 255).astype("uint8")
print("\n Converted to :-")
print("\n New diff ->",diff)
```

```python
"""The cv2.boundingRect() function of OpenCV is used to draw an approximate rectangle around the binary image."""
thresh = cv2.threshold(diff, 0, 255, cv2.THRESH_BINARY_INV | cv2.THRESH_OTSU)[1]

conts = cv2.findContours(thresh.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
conts = imutils.grab_contours(conts)

for c in conts:
    (x, y, w, h) = cv2.boundingRect(c)
    cv2.rectangle(imageA, (x,y), (x + w, y + h), (150, 150, 150), 2)
    cv2.rectangle(imageB, (x,y), (x + w, y + h), (150, 150, 150), 2)

fig, ax = plt.subplots(2,2, figsize=(10,10))
ax[0,0].imshow(imageA)
ax[0,0].set_title("Original")
ax[0,1].imshow(imageB)
ax[0,1].set_title("Modified")
ax[1,0].imshow(diff, cmap="gray")
ax[1,0].set_title("Difference")
ax[1,1].imshow(thresh, cmap="gray")
ax[1,1].set_title("Threshold")
plt.show()
```
