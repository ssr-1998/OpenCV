## High Pass Filtering 

There are **2 types of filtering for detecting an image edges :-**<br>
**1.** **High Pass Filtering**<br>
**2.** **Low Pass Filtering**<br><br>
Here we discussed **High Pass Filtering** which means that we will simply use **sobel_x / sobel_y** to get the edges of an object from an image.<br>
To see **Low Pass Filtering** please open my  **Low Pass Filtering (How To Detect Edges -2) Notebook.**


```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from matplotlib.pyplot import rcParams
```

```python
%matplotlib inline
rcParams["figure.figsize"] = 10,10
```

First we will load the image and will convert it to RGB channel from BGR channel.

```python
img = cv2.imread("images/curved_lane.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

Now, will convert it to gray scale, as image manipulation & pre-processing becomes easy in gray scale.

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img, cmap="gray")
plt.show()
```

**sobel_y, sobel_x means a matrix whose total will be zero.** These matrices are sent to each matrix of image data to only highlight those matrices where there is a high frequency area, to detect edges. To use these **sobels** we will use **cv2.filter2D() function.**<br>
**In sobel_x** -  {for example :-<br>[[ -1, 0, 1],<br>[ -2, 0, 2],<br>[ -1, 0, 1]]} means that we will sum it's values in x-axis direction to make it's **sum zero.**<br>
**Whereas in sobel_y** {for example :-<br>[[-1,-2,-1],<br>[0,0,0],<br>[1,2,1]]} this means that we will sum this matrix values in y-axis direction to make it's **sum zero.**<br><br>Here we first applied **sobel_y** & then will apply **sobel_x** in this image's edge detection notebook but to see a better example please open my **Low Pass Filtering (How To Detect Edges -2) Notebook.**

```python
sobel_y = np.array([[-1,-2,-1], [0,0,0], [1,2,1]])

filtered_sobel_y = cv2.filter2D(gray_img, -1, sobel_y)

plt.imshow(filtered_sobel_y, cmap="gray")
plt.show()
```

Now we just need to use the **cv2.threshold() function** to see the edges of objects.

```python
retval, binary_sobel_y = cv2.threshold(filtered_sobel_y, 50, 255, cv2.THRESH_BINARY)
plt.imshow(binary_sobel_y, cmap="gray")
plt.show()
```

## Now let's use sobel_x

```python
sobel_x = np.array([[1,0,-1], [-2,0,2], [1,0,-1]])

filtered_sobel_x = cv2.filter2D(gray_img, -1, sobel_x)
plt.imshow(filtered_sobel_x, cmap="gray")
plt.show()
```

```python
retval, binary_sobel_x = cv2.threshold(filtered_sobel_x, 50, 255, cv2.THRESH_BINARY)
plt.imshow(binary_sobel_x, cmap="gray")
plt.show()
```
